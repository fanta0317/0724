```python
#import modules
import re
from nltk.corpus import wordnet

#building keyword list
list_words = ["hello","timings"]
list_syn={}
#列舉list_words的元素
for word in list_words:
  synonyms=[]
  #找同義詞
  for syn in wordnet.synsets(word):
    #列同義詞
    for lem in syn.lemmas():
      #「^」-->不符合
      #除了a-zA-Z0-9 \n\.其他用空格取代(去除不必要符號)
      #sub=>取代
      lem_name = re.sub("[^a-zA-Z0-9 \n\.]",' ',lem.name())
      #在list-synonyms中append len_name
      synonyms.append(lem_name)
  #把list改成set(集合)
  list_syn[word]=set(synonyms)
print(list_syn)

#building dictionary intents & keyword
keywords={}
keywords_dict={}
keywords["greet"]=[]
for synonyms in list(list_syn["hello"]):
  keywords["greet"].append(".*\\b"+synonyms+"\\b.*")
#defining a new key(timing) in keywords dictionary
keywords["timings"]=[]
for synonyms in list(list_syn["timings"]):
  keywords["timings"].append(".*\\b"+synonyms+"\\b.*")
for intent,keys in keywords.items():
  keywords_dict[intent]=re.compile("|".join(keys))
print(keywords_dict)
# Building a dictionary of responses
responses={'greet':'Hello! How can I help you?','timings':'We are open from 9AM to 5PM, Monday to Friday. We are closed on weekends and public holidays.','fallback':'I dont quite understand. Could you repeat that?',}
print ("Welcome to MyBank. How may I help you?")
# While loop to run the chatbot indefinetely
while (True):
    # Takes the user input and converts all characters to lowercase
    user_input = input().lower()
    # Defining the Chatbot's exit condition
    if user_input == 'quit':
        print ("Thank you for visiting.")
        break
    matched_intent = None
    for intent,pattern in keywords_dict.items():
        # Using the regular expression search function to look for keywords in user input
        if re.search(pattern, user_input):
            # if a keyword matches, select the corresponding intent from the keywords_dict dictionary
            matched_intent=intent
    # The fallback intent is selected by default
    key='fallback'
    if matched_intent in responses:
        # If a keyword matches, the fallback intent is replaced by the matched intent as the key for the responses dictionary
        key = matched_intent
    # The chatbot prints the response that matches the selected intent
    print (responses[key])


```

```python
from keras.layers.serialization import activation
#預備建立順序模型
from keras.models import Sequential
#使每個節點互連
from keras.layers import Dense

import pandas as pd
df=pd.read_excel("/content/example_CO2.xlsx")
#df.head(10)

#資料正規化=> num-min/max-min =>使數值介於0~1之間
train=(df-df.min())/(df.max()-df.min())

#建model
model=Sequential()

#疊hidden
#unit=幾個神經元
#input有幾個-->做連線
model.add(Dense(units=5,input_dim=2,activation="sigmoid"))

#疊輸出層
model.add(Dense(units=1,activation="sigmoid"))

#定義損失函數
model.compile(loss="mse")

#訓練1000回合
#訓練過程存在train_history
train_history=model.fit(x=train[["GDP","perCapita"]],y=train["CO2"],epochs=10000)

import matplotlib.pyplot as plt
training_loss=train_history.history["loss"]
eporch_count=range(1,len(training_loss)+1)#+1是因為(1,1001)才會重複1000次
plt.plot(eporch_count,training_loss,"r")
plt.ylabel("loss")
plt.xlabel("eporch")
```
